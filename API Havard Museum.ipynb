# Print all the names and birth places of all British persons listed in the API

```
total_pages=data['info']['totalrecords']//100
total_pages
```

```
j=0 # index for person
for i in range(total_pages+1):
  query = {'apikey': '835fd7f0-b0e4-11ea-ae32-5f604ecf6a3b', 
         "q":"culture:British", 
         'sortorder': 'desc',
         'size':100,
         'page':i}
  url = 'https://api.harvardartmuseums.org/person'
  response = requests.get(url, params=query)
  data = response.json()
  records = data['records']
  print('Page: ',i)

  for record in records:
    #print(record)
    j+=1
    print(j,'. ',record['displayname'],': ', record['birthplace'])
```
